# Organize data

```python
# carbon_CTD: Creating new dataframe from CSV file, converting date to standardized YMD. 
carbon_CTD['date_ymd'] = pd.to_datetime(carbon_CTD['DATE']).dt.strftime('%Y%m%d') 
Doc_Nutrients['date_ymd'] = Doc_Nutrients['DATE'].astype(str) 
HPLC_Pig['Date_ymd'] = HPLC_Pig['Date'].astype(str) 

#Merging the Co2CTD and DocNutsCTD dataframes together using a left join. Created a new file using this merge. 
carbon_CTD = carbon_CTD.merge(Doc_Nutrients, left_on=['date_ymd', 'STATION', 'BOTTLE'], right_on=['date_ymd', 'STATION', 'BOTTLE'], how='left') #print(Co2_CTD) 
#Co2_CTD.to_csv("C:\TRACERS_SOphy\Output_Tracers\doc+co2.csv") 

#----------------------------------------------------------------------- 
#Dropping duplicate columns and renaming certain columns like date, depth, and latitude. 
#print(Co2_CTD.dtypes) 
#Co2_CTD=Co2_CTD.drop(Co2_CTD.columns[["DEPTH_x", "LATITUDE_x", "LONGITUDE_x", "SALINITY_CTD", "TEMP_CTD"]], axis=1) 
carbon_CTD.drop(columns=["DEPTH_y", "LATITUDE_y", "LONGITUDE_y", "SALINITY_CTD", "TEMP_CTD", "DATE_x" , "DATE_y"], inplace=True)

carbon_CTD = carbon_CTD.rename(columns={ 'date_ymd': 'Date', 'DEPTH_x' : 'Depth', 'LATITUDE_x' : 'LATITUDE', 'LONGITUDE_x' : 'LONGITUDE'}) 
#Co2_CTD.to_csv("C:\TRACERS_SOphy\Output_Tracers\doc+co2.csv") 


#----------------------------------------------------------------------- 
#Trying to merge the Co2CTD and DocNuts merged dataframe with the HPLC_Pig dataframe. 

#Loaded the HPLC_Pig file into a dataframe and then converted the Date into a string to allow for matching. 


HPLC_Pig = HPLC_Pig.merge(carbon_CTD, left_on=['Date_ymd', 'Depth'], right_on=['Date', 'Depth'], how='left') 

#HPLC_Pig.to_csv("C:\TRACERS_SOphy\Output_Tracers\HPLC_Pig+co2+doc.csv") 

#Due to the lack of shared columns between HPLC Pig and Co2CTD, its difficult to do a comprehensive merge. Only the data value is shared.
```
